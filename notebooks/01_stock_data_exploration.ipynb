# Notebook 01: Stock Data Exploration

**Objective:**  
This notebook explores daily stock price data using the Yahoo Finance API (`yfinance` library).  
We’ll collect and visualize historical stock price trends as the baseline for analyzing how market news and sentiment may affect stock performance.

**Data Source:**  
- Yahoo Finance via `yfinance` Python package  
- Daily stock prices include Open, High, Low, Close, Adj Close, and Volume  
- Example ticker: Apple (AAPL)

**Goal of this stage:**  
1. Retrieve 2 years of Apple’s historical stock prices.  
2. Clean and inspect the data.  
3. Visualize the price trend and trading volume over time.

```python
# Import core libraries
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
import seaborn as sns

# Make plots look cleaner
sns.set(style="whitegrid", palette="muted", font_scale=1.2)

print("✅ Libraries successfully imported.")
```

```python
# Download 2 years of historical data for Apple (AAPL)
ticker = "AAPL"
data = yf.download(ticker, period="2y")

# Display the first 5 rows
data.head()
```

```python
# Basic dataset overview
print("Shape of dataset:", data.shape)
print("\nMissing values per column:\n", data.isna().sum())

# Display summary statistics
data.describe()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(data.index, data["Close"], label="Close Price", linewidth=2)
plt.title(f"{ticker} Stock Closing Price (Last 2 Years)")
plt.xlabel("Date")
plt.ylabel("Price (USD)")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 4))
plt.plot(data.index, data["Volume"], color="skyblue", linewidth=1.5)
plt.title(f"{ticker} Daily Trading Volume (Last 2 Years)")
plt.xlabel("Date")
plt.ylabel("Volume")
plt.tight_layout()
plt.show()
```

### Interpretation: Daily Trading Volume Trends

The line chart above shows Apple's daily trading volume over the last two years.  
Several insights can already be observed:

1. **Spikes in Volume:** Noticeable peaks in trading activity often coincide with major company announcements, earnings releases, or market-wide events.  
2. **Investor Attention Patterns:** Sustained increases in trading volume may reflect heightened investor focus or sentiment — a key theme of this study.  
3. **Quiet Periods:** Periods of low volume can indicate decreased public attention or stability following major news cycles.

**Takeaway:**  
Trading volume can act as a *quantifiable proxy for investor attention*.  
Later, we’ll connect these volume spikes to specific news events and sentiment data to explore whether “attention” truly correlates with price behavior.
